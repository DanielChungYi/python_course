```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import locale
import matplotlib as mpl
from matplotlib.font_manager import FontProperties
font = FontProperties(fname=r"c:\windows\fonts\simsun.ttc",size=15)     
```

```python
# 新北市失業率—教育程度別 https://data.gov.tw/dataset/35960
df = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=35960&md5_url=a99f0ed3ab1ebca94bd53b0938f4ebec')
```

```python
df.head()
```

```python
df.columns = ['年','總計男','總計女','國小以下男','國小以下女','國中_初職_男','國中_初職_女','高中男','高中女','高職男','高職女','專科男','專科女','大學男','大學女','研究所男','研究所女']
```

```python
df.head()
```

```python
college_boy= df.loc[:,'大學男'].values
college_boy= np.delete(college_boy,-1, axis=0)
college_girl = df['大學女'].values
college_girl= np.delete(college_girl,-1, axis=0)
years = df.loc[:,'年'].values
years= np.delete(years,-1, axis=0)
for i in years:
    i = int(round(i))
ax = plt.subplot()
x = years
y = college_boy
z = college_girl
data1 = ax.bar(x-0.3, y,width=0.2,color='b',align='center')
data2 = ax.bar(x, z,width=0.2,color='r',align='center')
#ax.xaxis_date()
ax.legend( (data1[0], data2[0]), ('male', 'female'))
ax.autoscale(tight=True)
ind = np.arange(2006,2018,1)
width = 0.8 
ax.set_ylabel('失業率', fontproperties = font)
ax.set_xticks(ind)
plt.title('新北市大學畢業失業率', fontproperties = font)
plt.show()
```
